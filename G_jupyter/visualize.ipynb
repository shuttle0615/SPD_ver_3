```python
%cd /home/kyuholee/SPD_ver_3
from omegaconf import OmegaConf
import numpy as np

from D_train.iii_visualize import visualize
from B_data.ii_db_management import sort_coin_by_size

name = 'exp_5'
n = 5
coin_ls = ['QTUMUSDT'] #sort_coin_by_size()[:20]

data_config = OmegaConf.load('/home/kyuholee/SPD_ver_3/A_config/data_config.yaml')


prof_ls = []
win_rate_ls = []
for coin in coin_ls:
    prof, win_rate = visualize(data_config, name, n, coin)
    prof_ls.append(prof)
    if win_rate is None:
        continue
    else:
        win_rate_ls.append(win_rate)
    
print(f'mean and variacne of profit - mean: {np.mean(prof_ls)}, variance: {np.var(prof_ls)}')
print(f'mean and variacne of winrate - mean: {np.mean(win_rate_ls)}, variance: {np.var(win_rate_ls)}')

#print(sort_coin_by_size()[:40])
```
